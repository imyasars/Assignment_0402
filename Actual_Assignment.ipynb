```python
import pandas as pd
```

```python
data = pd.read_csv('sample.csv')
```

One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
test = data['text']
test
```

```python
encoder = OneHotEncoder()
```

```python
one_hot_encoded = encoder.fit_transform([[category] for category in test]).toarray()
print("Original categories:", test)
print("One-hot encoded result:")
print(one_hot_encoded)
```

BOW - Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
BOW=CountVectorizer()
```

```python
document_matrix=BOW.fit_transform(data["text"])
BOW.vocabulary_
```

```python
document_matrix[0].toarray()
```

N- Gram

Two-Gram

```python
bigram=CountVectorizer(ngram_range=(2,2))
bigramvocab=bigram.fit_transform(data["text"])
bigram.vocabulary_
```

3 - Gram

```python
bigram=CountVectorizer(ngram_range=(3,3))
bigramvocab=bigram.fit_transform(data["text"])
bigram.vocabulary_
```

Four - Gram

```python
bigram=CountVectorizer(ngram_range=(4,4))
bigramvocab=bigram.fit_transform(data["text"])
bigram.vocabulary_
```

TF - IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
```

```python
tfidf.fit_transform(data["text"]).toarray()
tfidf.get_feature_names_out()
tfidf.idf_
```
